Data Interpretation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=solar_weather['month'], y=solar_weather['Energy delta[Wh]'])
plt.title('Chart 01 - Energy based on Month')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.ylabel('Energy')

The lowest energy producing months are January, November and December.

The Highest energy producing month is June and besides that April, May, July and August are nicely energy producing months.
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=solar_weather['month'], y=solar_weather['GHI'],linewidth=5,errorbar=None)
sns.barplot(x=solar_weather['month'], y=solar_weather['GHI'], errwidth=0)
plt.title('Chart 02 - Solar Radiation based on Month')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.ylabel('Solar Radiation')
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x=solar_weather['month'], hue=solar_weather['weather_type'], palette='coolwarm')
plt.title('Chart 03 - Weather Type vs Month')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.ylabel('Count')
```
