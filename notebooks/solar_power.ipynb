```python
import numpy as np

# Filter for specific plant
plant_data = generation_data[generation_data['PLANT_ID'] == 4136001]

# Extract hour from datetime
plant_data['hour_of_day'] = pd.to_datetime(plant_data['DATE_TIME']).dt.hour

# Create temporary columns with zeros replaced by NaN for calculations
plant_data['DC_POWER_clean'] = plant_data['DC_POWER'].replace(0, np.nan)
plant_data['AC_POWER_clean'] = plant_data['AC_POWER'].replace(0, np.nan)

# Calculate hourly aggregates
avg_dc_power_hourly=plant_data.groupby('hour_of_day')['DC_POWER'].mean()
avg_ac_power_hourly=plant_data.groupby('hour_of_day')['AC_POWER'].mean()
avg_dc_clean=plant_data.groupby('hour_of_day')['DC_POWER_clean'].mean()
avg_ac_clean=plant_data.groupby('hour_of_day')['AC_POWER_clean'].mean()

# Calculate efficiency using clean averages (excluding zeros))
inv_efficiency = round((avg_ac_clean / avg_dc_clean) * 100, 2).fillna(0)

# Display efficiency by hour
print(inv_efficiency)


hourly_effi_result = pd.DataFrame({
    'avg_dc_power': avg_dc_power_hourly.values,
    'avg_ac_power': avg_ac_power_hourly.values,
    'pct_inverter_efficiency': round((avg_ac_clean / avg_dc_clean) * 100, 2).fillna(0)
}, index=avg_dc_clean.index).reset_index()[['hour_of_day', 'avg_dc_power', 'avg_ac_power', 'pct_inverter_efficiency']]

hourly_effi_result = hourly_effi_result.sort_values('hour_of_day')

print(hourly_effi_result)
```

```python
import pandas as pd

# Convert DATE_TIME to datetime format (add this at the beginning)
generation_data_clean['DATE_TIME'] = pd.to_datetime(generation_data_clean['DATE_TIME'])
weather_sensor_data['DATE_TIME'] = pd.to_datetime(weather_sensor_data['DATE_TIME'])

# Now your existing code will work
generation_data_clean['HOUR'] = generation_data_clean['DATE_TIME'].dt.hour
avg_power_clean = generation_data_clean.groupby('PLANT_ID')[['DC_POWER', 'AC_POWER']].mean()
gen_efficiency_clean = (avg_power_clean['AC_POWER'] / avg_power_clean['DC_POWER']) * 100

# Part 1: Generation data by PLANT_ID and hour (matching SQL subquery 'a')
gen_hourly = generation_data_clean.groupby(['PLANT_ID', 'HOUR']).agg(
    avg_dc_power=('DC_POWER', 'mean'),
    avg_ac_power=('AC_POWER', 'mean')
).reset_index()

# Calculate efficiency by hour (excluding zeros)
def calc_efficiency(group):
    dc_clean = group[group['DC_POWER'] != 0]['DC_POWER'].mean()
    ac_clean = group[group['AC_POWER'] != 0]['AC_POWER'].mean()
    if dc_clean and dc_clean != 0:
        return round((ac_clean / dc_clean) * 100, 2)
    return 0

gen_hourly['inverter_efficiency'] = generation_data_clean.groupby(['PLANT_ID', 'HOUR']).apply(calc_efficiency).values

# Part 2: Weather data by PLANT_ID and hour (matching SQL subquery 'b')
weather_sensor_data['HOUR'] = weather_sensor_data['DATE_TIME'].dt.hour

weather_hourly = weather_sensor_data.groupby(['PLANT_ID', 'HOUR']).agg(
    avg_amb_temp=('AMBIENT_TEMPERATURE', 'mean'),
    avg_mod_temp=('MODULE_TEMPERATURE', 'mean'),
    avg_irradiation=('IRRADIATION', 'mean')
).reset_index()

# Add temperature difference
weather_hourly['avg_temp_diff'] = weather_hourly['avg_mod_temp'] - weather_hourly['avg_amb_temp']

# Part 3: LEFT JOIN (same as SQL)
final_result = pd.merge(
    gen_hourly,
    weather_hourly,
    on=['PLANT_ID', 'HOUR'],
    how='left'
)

# Round numeric columns
final_result = final_result.round(2)

# Order by PLANT_ID and HOUR (same as SQL ORDER BY)
final_result = final_result.sort_values(['PLANT_ID', 'HOUR']).reset_index(drop=True)

# Rename HOUR to hour_of_day to match SQL output
final_result = final_result.rename(columns={'HOUR': 'hour_of_day'})

print(final_result)
```

```python
import pandas as pd
import os

# Convert DATE_TIME to datetime format - YOUR DATA USES DAY-MONTH-YEAR format
# Format: "15-05-2020 00:00" is day-month-year
generation_data_clean['DATE_TIME'] = pd.to_datetime(generation_data_clean['DATE_TIME'], format='%d-%m-%Y %H:%M')
weather_data['DATE_TIME'] = pd.to_datetime(weather_data['DATE_TIME'], format='%d-%m-%Y %H:%M')

# Now your existing code will work
generation_data_clean['HOUR'] = generation_data_clean['DATE_TIME'].dt.hour
avg_power_clean = generation_data_clean.groupby('PLANT_ID')[['DC_POWER', 'AC_POWER']].mean()
gen_efficiency_clean = (avg_power_clean['AC_POWER'] / avg_power_clean['DC_POWER']) * 100

# Part 1: Generation data by PLANT_ID and hour (matching SQL subquery 'a')
gen_hourly = generation_data_clean.groupby(['PLANT_ID', 'HOUR']).agg(
    avg_dc_power=('DC_POWER', 'mean'),
    avg_ac_power=('AC_POWER', 'mean')
).reset_index()

# Calculate efficiency by hour (excluding zeros)
def calc_efficiency(group):
    dc_clean = group[group['DC_POWER'] != 0]['DC_POWER'].mean()
    ac_clean = group[group['AC_POWER'] != 0]['AC_POWER'].mean()
    if dc_clean and dc_clean != 0:
        return round((ac_clean / dc_clean) * 100, 2)
    return 0

gen_hourly['inverter_efficiency'] = generation_data_clean.groupby(['PLANT_ID', 'HOUR']).apply(calc_efficiency).values

# Part 2: Weather data by PLANT_ID and hour (matching SQL subquery 'b')
weather_data['HOUR'] = weather_data['DATE_TIME'].dt.hour

weather_hourly = weather_data.groupby(['PLANT_ID', 'HOUR']).agg(
    avg_amb_temp=('AMBIENT_TEMPERATURE', 'mean'),
    avg_mod_temp=('MODULE_TEMPERATURE', 'mean'),
    avg_irradiation=('IRRADIATION', 'mean')
).reset_index()

# Add temperature difference
weather_hourly['avg_temp_diff'] = weather_hourly['avg_mod_temp'] - weather_hourly['avg_amb_temp']

# Part 3: LEFT JOIN (same as SQL)
final_result = pd.merge(
    gen_hourly,
    weather_hourly,
    on=['PLANT_ID', 'HOUR'],
    how='left'
)
# Round numeric columns
final_result = final_result.round(2)

# Order by PLANT_ID and HOUR (same as SQL ORDER BY)
final_result = final_result.sort_values(['PLANT_ID', 'HOUR']).reset_index(drop=True)

# Rename HOUR to hour_of_day to match SQL output
final_result = final_result.rename(columns={'HOUR': 'hour_of_day'})

# Display the result
print(final_result)

# Save to Excel file
excel_filename = 'plant_hourly_data.xlsx'

# Create Excel file with formatted table
with pd.ExcelWriter(excel_filename, engine='openpyxl') as writer:
    final_result.to_excel(writer, index=False, sheet_name='Plant Hourly Data')
    
    # Auto-adjust column widths
    worksheet = writer.sheets['Plant Hourly Data']
    for column in worksheet.columns:
        max_length = 0
        column_letter = column[0].column_letter
        for cell in column:
            try:
                if len(str(cell.value)) > max_length:
                    max_length = len(str(cell.value))
            except:
                pass
        adjusted_width = min(max_length + 2, 30)
        worksheet.column_dimensions[column_letter].width = adjusted_width

print(f"\n✅ Data successfully saved to '{excel_filename}'")
print(f"📊 You can now open this file in Microsoft Excel")
print(f"📍 File location: {os.path.abspath(excel_filename)}")
```
