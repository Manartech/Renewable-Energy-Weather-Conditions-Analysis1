```python
import matplotlib.pyplot as plt
import numpy as np

Energy_vs_weather = solar_weather.groupby('weather_type')['Energy delta[Wh]'].mean().reset_index()

fig, ax = plt.subplots(figsize=(19, 10))

# Set background color
fig.patch.set_facecolor('#f0f0f2')  # Figure background (around the plot)
ax.set_facecolor('#f0f0f2')          # Plot area background

# More vibrant colors
colors = ['#FF6B6B', '#4ECDC4', '#FFE66D', '#95E77E', '#C77DFF']
bars = ax.bar(Energy_vs_weather['weather_type'].astype(str), 
              Energy_vs_weather['Energy delta[Wh]'], 
              color=colors, 
              edgecolor='black', 
              linewidth=1.5)

# Add data labels with larger, bold text
ax.bar_label(bars, fmt='%.0f Wh', fontsize=25, fontweight='bold', padding=5)

# Custom legend for weather types
legend_labels = ['1: Clear', '2: Cloudy', '3: Lightly Rainy', 
                 '4: Heavy Rain', '5: Very Heavy Rain']
legend_elements = [plt.Rectangle((0,0), 1, 1, facecolor=colors[i], 
                    edgecolor='black', linewidth=1.5, label=legend_labels[i]) 
                   for i in range(5)]
ax.legend(handles=legend_elements, title="Weather Type Legend", 
          title_fontsize=15, fontsize=13, loc='upper right', framealpha=0.9, edgecolor='black')
# Custom legend for weather types with larger size
legend_labels = ['1: Clear', '2: Cloudy', '3: Lightly Rainy', 
                 '4: Heavy Rain', '5: Very Heavy Rain']
legend_elements = [plt.Rectangle((0,0), 1, 1, facecolor=colors[i], 
                    edgecolor='black', linewidth=1.5, label=legend_labels[i]) 
                   for i in range(5)]
ax.legend(handles=legend_elements, title="Weather Type Legend", 
          title_fontsize=20,           # Increased from 15
          fontsize=18,                 # Increased from 13
          loc='upper right', 
          framealpha=0.9, 
          edgecolor='black',
          handlelength=1,            # Longer color boxes
          handleheight=1,            # Taller color boxes
          borderpad=1.5)               # More padding inside legend

# Make legend bold
legend = ax.get_legend()
legend.get_title().set_fontweight('bold')
legend.get_title().set_fontsize(20)    # Explicitly set title size again
for text in legend.get_texts():
    text.set_fontweight('bold')
    text.set_fontsize(18)              # Explicitly set text size again
# Make legend bold
legend = ax.get_legend()
legend.get_title().set_fontweight('bold')
for text in legend.get_texts():
    text.set_fontweight('bold')

# Axis labels with larger, bold font
ax.set_xlabel("Weather Type", fontsize=30, fontweight='bold', labelpad=10)
ax.set_ylabel("Average Energy Watt", fontsize=30, fontweight='bold', labelpad=10)
# ax.set_title("Average Energy Production by Weather Type", fontsize=30, 
#              fontweight='bold', pad=15)

# Make axis tick labels larger and bold
ax.tick_params(axis='both', which='major', labelsize=20, width=1.5)
for label in ax.get_xticklabels():
    label.set_fontweight('bold')
for label in ax.get_yticklabels():
    label.set_fontweight('bold')

# Add grid for better readability (optional - keeps background visible)
# ax.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.8)
ax.set_axisbelow(True)

# Adjust y-axis to start from 0 or appropriate minimum
ax.set_ylim(bottom=0)

# Remove spines if you want a cleaner look (optional)
# for spine in ax.spines.values():
#     spine.set_visible(False)

plt.tight_layout()
plt.savefig("average_energy_prod_by_weather.png")
plt.savefig('average_energy_prod_by_weather.pdf')
plt.show()
```

```python
month_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57', '#FF9FF3', 
                '#54A0FF', '#5F27CD', '#00D2D3', '#FF9F43', '#00B894', '#E17055']

plt.figure(figsize=(8,7))
sns.barplot(data=solar_weather, x='month', y='Energy delta[Wh]', palette=month_colors)

# Bigger labels
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().annotate(f'{height:.0f} Wh', 
                       (p.get_x() + p.get_width()/2., height),
                       ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('Energy based on Month')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.ylabel('Energy')
plt.tight_layout()
plt.show()
# The lowest energy producing months are January, November and December.

# The Highest energy producing month is June and besides that April, May, July and August are nicely energy producing months.
```

```python
# Single green color for all bars
plt.figure(figsize=(40,15))
sns.barplot(data=solar_weather, x='hour', y='Energy delta[Wh]', color='#d5e590')

# Data labels - only for hours 7 through 14 (no lines)
for p in plt.gca().patches:
    height = p.get_height()
    x_pos = p.get_x() + p.get_width()/2.
    
    hour_value = len([patch for patch in plt.gca().patches if patch.get_x() < p.get_x()])
    
    if 7 <= hour_value <= 14:
        # Simple annotate without any arrow or line parameters
        plt.gca().annotate(f'{height:.0f} Wh', 
                           xy=(x_pos, height),
                           ha='center', va='bottom', fontsize=25, fontweight='bold')

plt.title('Average Energy by Hour (24h trend)', fontweight='bold', fontsize=20)
plt.xlabel('Hour of Day', fontweight='bold', fontsize=30)
plt.xticks(rotation=90, fontweight='bold', fontsize=30)
plt.yticks(fontweight='bold', fontsize=30)
plt.ylabel('Average Energy delta (Wh)', fontweight='bold', fontsize=30)
plt.tight_layout()
plt.show()
```

```python
# Load your data (assuming you already have it loaded as solar_weather)
# solar_weather = pd.read_csv('solar_weather.csv')

# Ensure Time is datetime
solar_weather['Time'] = pd.to_datetime(solar_weather['Time'])

# Create month and year columns for grouping
solar_weather['month'] = solar_weather['Time'].dt.month
solar_weather['year'] = solar_weather['Time'].dt.year
solar_weather['month_name'] = solar_weather['Time'].dt.strftime('%B')

# 1. DAILY LOAD PROFILE (Hourly average across all months)
daily_load = solar_weather.groupby('hour')['Energy delta[Wh]'].mean().reset_index()
daily_load.columns = ['hour', 'avg_energy_wh']

print("="*60)
print("DAILY LOAD PROFILE (Average across all data)")
print("="*60)
print(daily_load.to_string(index=False))

# 2. MONTHLY LOAD PROFILE (Hourly average by month)
monthly_load = solar_weather.groupby(['month', 'hour'])['Energy delta[Wh]'].mean().reset_index()
monthly_load.columns = ['month', 'hour', 'avg_energy_wh']

# Add month names for better readability
month_names = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',
               7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
monthly_load['month_name'] = monthly_load['month'].map(month_names)

print("\n" + "="*60)
print("MONTHLY LOAD PROFILE (Hourly averages by month)")
print("="*60)
# Pivot for better visualization
monthly_pivot = monthly_load.pivot(index='hour', columns='month_name', values='avg_energy_wh')
print(monthly_pivot.round(2).to_string())

# 3. HIGH SOLAR HOURS (GHI > 200) - Load during good solar conditions
high_solar_data = solar_weather[solar_weather['GHI'] > 200].copy()
high_solar_load = high_solar_data.groupby('hour')['Energy delta[Wh]'].mean().reset_index()
high_solar_load.columns = ['hour', 'avg_energy_during_high_solar']

# Also by month for high solar conditions
high_solar_monthly = high_solar_data.groupby(['month', 'hour'])['Energy delta[Wh]'].mean().reset_index()
high_solar_monthly.columns = ['month', 'hour', 'avg_energy_during_high_solar']
high_solar_monthly['month_name'] = high_solar_monthly['month'].map(month_names)

print("\n" + "="*60)
print("LOAD DURING HIGH SOLAR CONDITIONS (GHI > 200 W/m²)")
print("="*60)
print(high_solar_load.to_string(index=False))

# 4. COMPARISON: Load during high solar vs average
comparison = pd.merge(daily_load, high_solar_load, on='hour', how='left')
comparison['difference'] = comparison['avg_energy_wh'] - comparison['avg_energy_during_high_solar']
comparison['percent_difference'] = (comparison['difference'] / comparison['avg_energy_wh'] * 100).round(1)

print("\n" + "="*60)
print("COMPARISON: Average Load vs Load During High Solar Hours")
print("="*60)
print(comparison.to_string(index=False))

# 5. SEASONAL SOLAR POTENTIAL ANALYSIS
# Calculate average GHI by hour and month
solar_potential = solar_weather.groupby(['month', 'hour'])['GHI'].mean().reset_index()
solar_potential.columns = ['month', 'hour', 'avg_ghi']
solar_potential['month_name'] = solar_potential['month'].map(month_names)
solar_pivot = solar_potential.pivot(index='hour', columns='month_name', values='avg_ghi')

print("\n" + "="*60)
print("SOLAR POTENTIAL (Average GHI by Hour and Month)")
print("="*60)
print(solar_pivot.round(2).to_string())

# 6. IDENTIFY OPTIMAL HOURS FOR SOLAR BY SEASON
print("\n" + "="*60)
print("OPTIMAL SOLAR HOURS BY SEASON")
print("="*60)

seasons = {
    'Winter (Dec-Feb)': [12, 1, 2],
    'Spring (Mar-May)': [3, 4, 5],
    'Summer (Jun-Aug)': [6, 7, 8],
    'Fall (Sep-Nov)': [9, 10, 11]
}

for season, months in seasons.items():
    season_data = solar_weather[solar_weather['month'].isin(months)]
    # Find top 5 hours for solar generation
    top_solar_hours = season_data.groupby('hour')['GHI'].mean().sort_values(ascending=False).head(5)
    print(f"\n{season}:")
    print(f"  Best hours for solar: {list(top_solar_hours.index)}")
    print(f"  Peak GHI: {top_solar_hours.iloc[0]:.1f} W/m²")
    
    # Load during those hours
    peak_hour = top_solar_hours.index[0]
    load_at_peak = season_data[season_data['hour'] == peak_hour]['Energy delta[Wh]'].mean()
    print(f"  Average load at peak solar hour ({peak_hour}:00): {load_at_peak:.0f} Wh")

# 7. YEARLY TREND (2017-2022)
print("\n" + "="*60)
print("YEARLY TREND ANALYSIS (2017-2022)")
print("="*60)

# Group by year and month
yearly_monthly = solar_weather.groupby(['year', 'month'])['Energy delta[Wh]'].sum().reset_index()
yearly_monthly['month_name'] = yearly_monthly['month'].map(month_names)

# Pivot for yearly comparison
yearly_pivot = yearly_monthly.pivot(index='year', columns='month_name', values='Energy delta[Wh]')
print("\nTotal Energy Consumption by Year and Month (kWh):")
print((yearly_pivot / 1000).round(2).to_string())

# Yearly solar potential
yearly_solar = solar_weather.groupby(['year', 'month'])['GHI'].mean().reset_index()
yearly_solar['month_name'] = yearly_solar['month'].map(month_names)
solar_yearly_pivot = yearly_solar.pivot(index='year', columns='month_name', values='GHI')

print("\nAverage GHI by Year and Month (W/m²):")
print(solar_yearly_pivot.round(2).to_string())

# 8. VISUALIZATION
fig, axes = plt.subplots(3, 2, figsize=(15, 12))

# Plot 1: Daily load profile
axes[0, 0].plot(daily_load['hour'], daily_load['avg_energy_wh'], 'b-', linewidth=2, marker='o')
axes[0, 0].set_xlabel('Hour of Day')
axes[0, 0].set_ylabel('Energy (Wh)')
axes[0, 0].set_title('Average Daily Load Profile (All Months)')
axes[0, 0].grid(True)
axes[0, 0].axvline(x=10, color='r', linestyle='--', alpha=0.5)

# Plot 2: Heatmap of monthly load
load_pivot = monthly_load.pivot(index='hour', columns='month', values='avg_energy_wh')
im = axes[0, 1].imshow(load_pivot.values, aspect='auto', cmap='YlOrRd', 
                       extent=[1, 12, 23, 0])
axes[0, 1].set_xlabel('Month')
axes[0, 1].set_ylabel('Hour of Day')
axes[0, 1].set_title('Monthly Load Heatmap (Wh)')
plt.colorbar(im, ax=axes[0, 1])

# Plot 3: Solar potential heatmap
solar_pivot_values = solar_potential.pivot(index='hour', columns='month', values='avg_ghi')
im2 = axes[1, 0].imshow(solar_pivot_values.values, aspect='auto', cmap='YlGn', 
                        extent=[1, 12, 23, 0])
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Hour of Day')
axes[1, 0].set_title('Solar Potential Heatmap (GHI W/m²)')
plt.colorbar(im2, ax=axes[1, 0])

# Plot 4: Load during high solar vs average
axes[1, 1].plot(comparison['hour'], comparison['avg_energy_wh'], 'b-', label='Average Load', linewidth=2)
axes[1, 1].plot(comparison['hour'], comparison['avg_energy_during_high_solar'], 'g--', 
                label='Load During High Solar (GHI>200)', linewidth=2)
axes[1, 1].set_xlabel('Hour of Day')
axes[1, 1].set_ylabel('Energy (Wh)')
axes[1, 1].set_title('Load Comparison: Average vs High Solar Conditions')
axes[1, 1].legend()
axes[1, 1].grid(True)

# Plot 5: Monthly energy consumption patterns
monthly_total = solar_weather.groupby('month')['Energy delta[Wh]'].sum().reset_index()
monthly_total['month_name'] = monthly_total['month'].map(month_names)
axes[2, 0].bar(monthly_total['month_name'], monthly_total['Energy delta[Wh]'] / 1000, color='skyblue')
axes[2, 0].set_xlabel('Month')
axes[2, 0].set_ylabel('Total Energy (kWh)')
axes[2, 0].set_title('Total Monthly Energy Consumption')
axes[2, 0].tick_params(axis='x', rotation=45)

# Plot 6: Scatter of GHI vs Energy by season
colors = {'Winter': 'blue', 'Spring': 'green', 'Summer': 'red', 'Fall': 'orange'}
for season, months in seasons.items():
    season_data = solar_weather[solar_weather['month'].isin(months)]
    axes[2, 1].scatter(season_data['GHI'], season_data['Energy delta[Wh]'], 
                      alpha=0.3, s=5, c=colors[season], label=season)
axes[2, 1].set_xlabel('GHI (W/m²)')
axes[2, 1].set_ylabel('Energy Delta (Wh)')
axes[2, 1].set_title('Energy Consumption vs Solar Radiation by Season')
axes[2, 1].legend()
axes[2, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 9. KEY STATISTICS SUMMARY
print("\n" + "="*60)
print("KEY STATISTICS SUMMARY")
print("="*60)

# Best month for solar
best_solar_month = solar_potential.groupby('month')['avg_ghi'].mean().idxmax()
print(f"Best month for solar (highest avg GHI): {month_names[best_solar_month]}")

# Worst month for solar
worst_solar_month = solar_potential.groupby('month')['avg_ghi'].mean().idxmin()
print(f"Worst month for solar (lowest avg GHI): {month_names[worst_solar_month]}")

# Month with highest energy consumption
highest_load_month = monthly_total.loc[monthly_total['Energy delta[Wh]'].idxmax()]
print(f"Month with highest energy consumption: {month_names[highest_load_month['month']]}")

# Best solar hour (overall)
best_solar_hour = solar_potential.groupby('hour')['avg_ghi'].mean().idxmax()
print(f"Best hour for solar (overall): {best_solar_hour}:00")

# Load during best solar hour
load_at_best_hour = daily_load[daily_load['hour'] == best_solar_hour]['avg_energy_wh'].values[0]
print(f"Average load during best solar hour: {load_at_best_hour:.0f} Wh")

# Solar contribution potential
print(f"\nSOLAR CONTRIBUTION POTENTIAL:")
print(f"If you install solar to offset load during peak solar hours (10:00-14:00):")
peak_solar_hours = [10, 11, 12, 13, 14]
peak_load = solar_weather[solar_weather['hour'].isin(peak_solar_hours)]['Energy delta[Wh]'].mean()
print(f"  Average load during peak solar window: {peak_load:.0f} Wh/hour")
print(f"  Required solar array size (assuming 1000 W/m² peak, 80% efficiency):")
print(f"  At average GHI of {solar_potential[solar_potential['hour'].isin(peak_solar_hours)]['avg_ghi'].mean():.0f} W/m²:")
required_size = peak_load / (solar_potential[solar_potential['hour'].isin(peak_solar_hours)]['avg_ghi'].mean() / 1000 * 0.8)
print(f"  Required array: {required_size:.0f} kWp")

# Export key results for further analysis
print("\n" + "="*60)
print("EXPORTING KEY RESULTS")
print("="*60)

# # Save to CSV for later use
# daily_load.to_csv('daily_load_profile.csv', index=False)
# monthly_load.to_csv('monthly_load_profile.csv', index=False)
# solar_potential.to_csv('solar_potential.csv', index=False)
# comparison.to_csv('load_comparison.csv', index=False)

# print("✓ Daily load profile saved to 'daily_load_profile.csv'")
# print("✓ Monthly load profile saved to 'monthly_load_profile.csv'")
# print("✓ Solar potential data saved to 'solar_potential.csv'")
# print("✓ Load comparison saved to 'load_comparison.csv'")
```

```python
# Calculate statistics
stats = solar_weather['Energy delta[Wh]'].describe()
print("Original Statistics:")
print(stats)
print(f"\nSkewness: {solar_weather['Energy delta[Wh]'].skew():.2f}")
print(f"Zero values: {(solar_weather['Energy delta[Wh]'] == 0).sum():,} ({(solar_weather['Energy delta[Wh]'] == 0).sum()/len(solar_weather)*100:.1f}%)")

# SIMPLE SKEW-ADAPTED IQR METHOD
# For highly skewed data (median = 0), use asymmetric bounds

Q1 = solar_weather['Energy delta[Wh]'].quantile(0.25)
Q3 = solar_weather['Energy delta[Wh]'].quantile(0.75)
IQR = Q3 - Q1

# ADJUSTED multipliers for positive skew
lower_multiplier = 1.5  # Keep standard for lower bound
upper_multiplier = 3.0  # Use 3× for upper bound (less aggressive)

lower_bound = Q1 - lower_multiplier * IQR
upper_bound = Q3 + upper_multiplier * IQR

print(f"\nSkew-Adapted IQR Bounds:")
print(f"  Lower bound: {lower_bound:.0f} Wh")
print(f"  Upper bound: {upper_bound:.0f} Wh")

# Detect outliers
outliers = solar_weather[(solar_weather['Energy delta[Wh]'] < lower_bound) | 
                          (solar_weather['Energy delta[Wh]'] > upper_bound)]

print(f"  Outliers detected: {len(outliers)} ({len(outliers)/len(solar_weather)*100:.2f}%)")

# Create cleaned dataset (cap outliers)
solar_weather_clean = solar_weather.copy()
solar_weather_clean['Energy delta_clean'] = solar_weather_clean['Energy delta[Wh]'].clip(
    lower=lower_bound, upper=upper_bound
)

# Compare before/after
print("\n" + "="*50)
print("BEFORE vs AFTER Outlier Capping")
print("="*50)
print(f"{'Metric':<15} {'Original':<15} {'Cleaned':<15}")
print("-"*45)
print(f"{'Mean':<15} {solar_weather['Energy delta[Wh]'].mean():<15.0f} {solar_weather_clean['Energy delta_clean'].mean():<15.0f}")
print(f"{'Std':<15} {solar_weather['Energy delta[Wh]'].std():<15.0f} {solar_weather_clean['Energy delta_clean'].std():<15.0f}")
print(f"{'Max':<15} {solar_weather['Energy delta[Wh]'].max():<15.0f} {solar_weather_clean['Energy delta_clean'].max():<15.0f}")
print(f"{'99th %ile':<15} {solar_weather['Energy delta[Wh]'].quantile(0.99):<15.0f} {solar_weather_clean['Energy delta_clean'].quantile(0.99):<15.0f}")
```
